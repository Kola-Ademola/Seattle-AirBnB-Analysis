```python
#import the neccesary libraries
import pandas as pd
import numpy as np

# display all columns in the dataset
pd.set_option('display.max_columns', None)
```

```python
#load the datasets for cleaning

listings_df = pd.read_csv('listings.csv')
calendar_df = pd.read_csv('calendar.csv')
```

```python
#view listings_df

listings_df.head()
```

```python
# I will be making a copy of the datasets before I start cleaning

listings_clean = listings_df.copy()
calendar_clean = calendar_df.copy()
```

```python
#getting the list of unwanted columns

listings_clean.columns
```

```python
#dropping irrelevant columns

#list of columns that will be dropped
cols = ['listing_url', 'scrape_id', 'last_scraped', 'name', 'summary',
       'space', 'description', 'experiences_offered', 'neighborhood_overview',
       'notes', 'transit', 'thumbnail_url', 'medium_url', 'picture_url',
       'xl_picture_url', 'host_url', 'host_since', 'host_is_superhost',
       'host_location', 'host_about', 'host_response_time', 'host_verifications',
       'host_response_rate', 'host_acceptance_rate',
       'host_thumbnail_url', 'host_picture_url', 'host_neighbourhood',
       'host_listings_count', 'host_has_profile_pic',
       'street', 'neighbourhood',
       'neighbourhood_group_cleansed', 'bathrooms',
       'smart_location', 'country_code', 'state', 'market',
       'is_location_exact', 'amenities', 'square_feet',
        'weekly_price', 'monthly_price', 'security_deposit',
       'cleaning_fee', 'guests_included', 'extra_people', 'minimum_nights',
       'maximum_nights', 'calendar_updated',
       'availability_30', 'availability_60', 'availability_90',
       'availability_365', 'calendar_last_scraped', 'number_of_reviews',
       'first_review', 'last_review','review_scores_location',
       'review_scores_accuracy', 'review_scores_cleanliness',
       'review_scores_checkin', 'review_scores_communication', 'review_scores_value',
        'requires_license','license', 'jurisdiction_names', 'require_guest_profile_picture',
       'require_guest_phone_verification', 'calculated_host_listings_count','reviews_per_month']

# dropping above columns

listings_clean = listings_clean.drop(cols, axis = 1)

# confirm changes

listings_clean.info()
```

```python
#view dataset
listings_clean.head()
```

```python
#checking for null values

listings_clean.isnull().sum()
```

```python
#dropping rows with null values

listings_clean = listings_clean.dropna()

#confirm there is no more null rows

listings_clean.isnull().sum()
```

```python
#removing the '$' sign from the 'price' column

listings_clean['price'] = listings_clean['price'].str.replace('$', '')

#confirm change
listings_clean['price']
```

```python
#checking the zipcodes

listings_clean['zipcode'].value_counts()
```

```python
#correcting the wrong zipcode
listings_clean['zipcode'] = listings_clean['zipcode'].str.replace('99\n98122', '98122')

#confirm change
listings_clean['zipcode'].value_counts()
```

```python
#converting the 'review_scores_rating' column to a 10-scale rating format
listings_clean['review_scores_rating'] = listings_clean['review_scores_rating'] / 10

#confirm change
listings_clean['review_scores_rating']
```

```python
# check the calender_df

calendar_clean.head()
```

```python
#removing the '$' sign from the 'price' column

calendar_clean['price'] = calendar_clean['price'].str.replace('$', '')

#confirm change
calendar_clean['price']
```

```python
#view datasets again
listings_clean.head(10)
```

```python
#view datasets again
calendar_clean.head(10)
```

## Saving & Exporting cleaned data

```python
#exporting the cleaned listings data
listings_clean.to_csv('listings_cleaned.csv', index = False)

#exporting the cleaned calendar data
calendar_clean.to_csv('calendar_cleaned.csv', index = False)
```
